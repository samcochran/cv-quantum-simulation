# Harmonic Oscillator Lattice: Phonons

```python
omega = .9
F = 1/2*omega**2*np.array([[2, -2, 0, -2],
                            [0, 2, -2, 0],
                            [0, 0, 2, -2],
                            [0, 0, 0, 2]])
K = 100
dt = .05
L = 8
xvec, pvec = np.linspace(-L, L, 500), 1*np.linspace(-L, L, 500)
eng = sf.Engine('gaussian')
x0 = [-1, .5, 1.3, -.9]
simulator = quadratic_simulator(F)

snapshots = [[] for _ in range(F.shape[0])]
progress = tqdm(total=K)
for k in range(K):
    results = simulator.run(k, dt, xvec, pvec, eng, x0=x0)
    for snapshot_list, result in zip(snapshots, results):
        snapshot_list.append(result)
    progress.update(1)
progress.close()
```
